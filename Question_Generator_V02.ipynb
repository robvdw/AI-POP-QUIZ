```python
quiz = '''

# Which of the following machine learning model needs the presence of a teacher?
type: mc 

* Unsupervised
-c f
-f false 

* Supervised
-c t
-f Correct

'''
```

```python
def md2json(md):
    quiz_types = {"multiple_choice": ["mc", "multiple_choice"],
                  "many_choice": ["mnc", "mmc", "many_choice"],
                 "number": ["number", "n"]
                    }
    x=0
    lines = md.splitlines()
    quiz_data = []
    for line in lines:
        line = line.strip()
        if line:
            if line[0] == "#":
                question_data = {}
                
                question = line[1:].strip()
                question_data["question"] = question
                #find type of quiz
                q = lines[x+1]
                
                for quiz in quiz_types:
                    for item in quiz_types[quiz]:
                        if item in q:
                            type_quiz = quiz
                answers = "\n".join(lines[x+2:]).split("#")[0]
                answers = answers.split("*")[1:]
                answer_data = []
                corrects= 0
                for answer in answers:
                    answer = answer.strip()
                    answer, correct, feedback = answer.splitlines()[:3]
                    if "f" in correct.lower():
                        correct = False
                    elif "t" in correct.lower():
                        correct = True
                    if "-f" == feedback[:2]:
                        feedback = feedback[2:].strip()
                    if correct == True:
                        corrects = corrects+1
                    answer_data.append(
                        {"answer": answer,
                        "correct": correct,
                         "feedback": feedback
                         })
                    if corrects == 1:
                        type_quiz = "multiple_choice"
                    elif corrects > 1:
                        type_quiz = "many_choice"
                quiz_data.append({"question": question,
                                 "type": type_quiz,
                               "answers": answer_data})
                
        x=x+1
    
    return json.dumps(quiz_data, indent=4)
myjson_quiz = md2json(quiz)
print (myjson_quiz)
```
